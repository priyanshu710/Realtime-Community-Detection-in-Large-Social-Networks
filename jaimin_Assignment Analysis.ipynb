```python
import networkx as nx
import random
import numpy as np
import pandas as pd
import community
import matplotlib.pyplot as plt
import community
```

```python
# Load the Email-Eu-Core dataset
G = nx.read_edgelist("C:\MTech\Second Semester\Web and Social Computing(IT752)\Project\Email-EuAll.txt")
```

```python
print("Total number ofnodes in the given graph: ", int(G.number_of_nodes())) 
print("Total number of edges in the given graph: ", int(G.number_of_edges()))
```

```python
#Average Degree
print("Average Degree of each node is:")
average_degree = nx.average_degree_connectivity(G)
print(average_degree)
```

```python
# Calculating degree distribution. 

def plot_degree_dist(graph):
    degrees = [graph.degree(n) for n in graph.nodes()]
    plt.hist(degrees,color = "black")
    plt.xlabel("Degree")
    plt.ylabel("Number of nodes")
    plt.title("Degree distribution in the graph")
    plt.show()

plot_degree_dist(G)

degrees = nx.degree(G)
all_degrees = list(dict(degrees).values())

unique_degrees = list(set(all_degrees))

count_degrees = []
for i in unique_degrees:
  count = all_degrees.count(i)
  count_degrees.append(count)
plt.xlabel("Degree")
plt.ylabel("Number of nodes")
plt.title("Degree distribution in the graph")
plt.plot(unique_degrees,count_degrees,color = "gray")
plt.show()
```

```python
 #Calculating Sparseness of the graph.
print("Sparseness of the graph:", nx.density(G))
```

```python
#Calculating diameter of the graph.

components = nx.connected_components(G)
largest_component = max(components, key=len)

subgraph = G.subgraph(largest_component)
diameter = nx.diameter(subgraph)
print("Network diameter of largest component:", diameter)
```
